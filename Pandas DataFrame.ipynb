```python
#Reading the ipl dataset
ipl=pd.read_csv('IPL Ball-by-Ball 2008-2020.csv')
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
ipl.head()
```

```python
ipl.shape
```

```python
#find the top10 batsman interms of run score
ipl.groupby('batsman')['batsman_runs'].sum().sort_values(ascending=False).head(10)
```

```python
#find the batsman with max no. of sixes
six=ipl[ipl['batsman_runs']==6]
six.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
#find the batsman with most number of six and four
temp_df=ipl[ipl['over']>15]
temp_df=temp_df[(temp_df['batsman_runs']==4)|(temp_df['batsman_runs']==6)]
temp_df.groupby('batsman')['batsman'].count().sort_values(ascending=False).head(1).index[0]
```

```python
#find the virat kholi records against all the teams
temp=ipl[ipl['batsman']=='V Kohli']
temp.groupby('bowling_team')['batsman_runs'].sum().reset_index()
```

```python
#Create the function that can return the highest score of any batsman
def highest(batsman):
    temp_df=ipl[ipl['batsman']==batsman]
    return temp_df.groupby('id')['batsman_runs'].sum().sort_values(ascending=False).values[0]
```

```python
highest('V Kohli')
```
